```python
import pandas as pd
```

```python
file_1 = "Resources/Accident_Information.csv"
accident_df = pd.read_csv(file_1)

accident_df.head()
```

```python
file_2 = "Resources/Vehicle_Information.csv"
vehicle_df = pd.read_csv(file_2)

vehicle_df.head()
```

```python
accident_df.info()
```

```python
accident_df.count()
```

```python
accident_df1=accident_df.dropna(how='any')
accident_df1
```

```python
vehicle_df.info()
```

```python
vehicle_df.count()
```

```python
vehicle_df1=vehicle_df.dropna(how='any')
vehicle_df1
```

```python
df_final = accident_df1.merge(vehicle_df1, on="Accident_Index", how = 'outer')
df_final
```

```python
df1_final=df_final.dropna(how='any')
df1_final
```

```python
df1_final.info()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
df1_final.to_csv("Resources/merged.csv", index=True, header=True)
```
